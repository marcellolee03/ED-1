**Exercício 1:**<br>
Implemente as classes ArrayStack, ArrayQueue e ArrayDeque

```python
# ArrayStack
class ArrayStack:
    def __init__(self, initial_stack = None):
        if initial_stack is None:
            self.stack = []
        else:
            self.stack = list(initial_stack)
    
    def is_empty(self):
        if len(self.stack) == 0:
            return True
        else:
            return False

    # Complexity O(1)
    def pop(self):
        if self.is_empty():
            raise IndexError('Pop from an empty stack')
        return self.stack.pop()
    
    # Complexity O(1)
    def push(self, value):
        self.stack.append(value)

    def peek(self):
        if self.is_empty():
            raise IndexError('Peek from an empty stack')
        return self.stack[-1]
    
    def __len__(self):
        return len(self.stack)
    
    def __str__(self):
        return str(self.stack)


# ArrayQueue
class ArrayQueue:
    def __init__(self, initial_queue = None):
        if initial_queue == None:
            self.queue = []
        else:
            self.queue = list(initial_queue)
    
    def is_empty(self):
        if len(self.queue) == 0:
            return True
        else:
            return False

    # Complexity O(1)
    def enqueue(self, value):
        self.queue.append(value)

    # Complexity O(n)
    def dequeue(self):
        if self.is_empty():
            raise IndexError('Dequeue from an empty queue')
        return self.queue.pop(0)
    
    def peek(self):
        if self.is_empty():
            raise IndexError('Peek into an empty queue')
        else:
            return self.queue[0]
    
    def __len__(self):
        return len(self.queue)
    
    def __str__(self):
        return str(self.queue)
    

# ArrayDeque
class ArrayDeque:
    def __init__(self, initial_deque = None):
        if initial_deque == None:
            self.deque = []
        else:
            self.deque = list(initial_deque)
    
    def is_empty(self):
        if len(self.deque) == 0:
            return True
        else:
            return False

    # Complexity O(1)
    def append(self, value):
        self.deque.append(value)
    
    # Complexity O(1)
    def pop(self):
        if self.is_empty():
            raise IndexError('Pop from empty deque')
        else:
            return self.deque.pop()

    # Complexity O(n)
    def append_left(self, value):
        self.deque.insert(0, value)

    # Complexity O(n)
    def pop_left(self):
        if self.is_empty():
            raise IndexError('Pop from empty deque')
        else:
            return self.deque.pop(0)
    
    def __len__(self):
        return len(self.deque)
    
    def __str__(self):
        return str(self.deque)
    
    def first(self):
        if self.is_empty():
            raise IndexError('Peek into an empty deque.')
        else:
            return self.deque[0]
        
    def last(self):
        if self.is_empty():
            raise IndexError('Peek into an empty deque.')
        else:
            return self.deque[len(self.deque) - 1]
```

**Exercício 3:**<br>
Implemente uma função com assinatura transfer(S, T) que transfira todos os elementos da
pilha S para a pilha T, de modo que o elemento que começa no topo de S seja o primeiro a
ser inserido em T, e o elemento na parte inferior de S termine no topo de T.

```python
def transfer(S, T):
    for item in range(0, len(S)):
        T.push(S.pop())

# Example
print('Example use: ')
S = ArrayStack([0, 1, 2, 3])
T = ArrayStack()
print(f'S array stack before transfer: {S}')
print(f'T array stack before transfer: {T}')

transfer(S, T)

print(f'S array stack after transfer: {S}')
print(f'T array stack after transfer: {T}')
```

**Exercício 5:**<br>
Implemente uma função que inverte uma lista de elementos colocando-os em uma pilha em uma ordem e escrevendo-os de volta na lista na ordem inversa.

```python
def list_inverter(L):
    aux = ArrayStack(L)
    inverted_list = []
    
    for i in range(0, len(aux)):
        inverted_list.append(aux.pop())
    
    L[:] = inverted_list

# Example
L = [0, 1, 2, 3]
print(f'L before list_inverter: {L}')
list_inverter(L)
print(f'L after list_inverter: {L}')
```

 **Exercício 8:**<br>
 Escreva um programa para verificar se em uma string contendo uma expressão aritmética, os parênteses de <br>
 abertura e fechamento estão bem formatados ou não.

```python
def are_parenthesis_valid(operation: str):
    aux = ArrayStack()
    try:
        for element in operation:
            if element == '(':
                aux.push(1)
            elif element == ')':
                aux.pop()
    except IndexError:
        return False
    
    return aux.is_empty()

# Example:
print(f'Result for (a+b)*)c(: {are_parenthesis_valid("(a+b)*)c(")}')
print(f'Result for ((a+b): {are_parenthesis_valid("((a+b)")}')
print(f'Result for (a*(b+c)): {are_parenthesis_valid("(a*(b+c))")}')
```

**Exercício 9:**<br>
Escreva um programa para converter uma expressão aritmética na forma prefixada para formas infixas e pós-fixadas equivalentes.

```python
def converter(expression: str):
    infix_aux = ArrayStack(expression)
    postfix_aux = ArrayStack(expression)

    infix = ArrayStack()
    postfix = ArrayStack()

    operators = ['+', '-', '*', '/']

    for i in range(0, len(expression)):
        obj = infix_aux.pop()
        
        if not obj in operators:
            infix.push(obj)
        else:
            if obj == '+' or obj == '-':
                operand1 = infix.pop()
                operand2 = infix.pop()

                infix.push(f'({operand1}{obj}{operand2})')
            else:
                operand1 = infix.pop()
                operand2 = infix.pop()

                infix.push(f'{operand1}{obj}{operand2}')
    
    for i in range(0, len(expression)):
        obj = postfix_aux.pop()

        if not obj in operators:
            postfix.push(obj)
        else:
            if obj == '+' or obj == '-':
                operand1 = postfix.pop()
                operand2 = postfix.pop()

                postfix.push(f'({operand1}{operand2}{obj})')
            else:
                operand1 = postfix.pop()
                operand2 = postfix.pop()

                postfix.push(f'{operand1}{operand2}{obj}')


    return {
        'infix': infix,
        'postfix': postfix
    }
    
result = converter('+-A*BCD')

print(result['infix'])
print(result['postfix'])
```

**Exercício 10:**<br>

```python
def calcultate(expression: list[str]):
    operators = ['+', '-', '*', '/']
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2}

    operators_stack = ArrayStack()
    postfixed_expression = ArrayStack()

    for token in expression:
        try:
            number = int(token)
            postfixed_expression.push(number)
        except ValueError:
            if token in operators:
                while (not operators_stack.is_empty() and 
                    operators_stack.peek() != "(" and
                    precedence.get(operators_stack.peek(), 0) >= precedence.get(token, 0)):
                    
                    postfixed_expression.push(operators_stack.pop())
                operators_stack.push(token)
            elif token == "(":
                operators_stack.push(token)
            elif token == ")":
                operator_token = operators_stack.pop()
                while not operator_token == "(":
                    postfixed_expression.push(operator_token)
                    operator_token = operators_stack.pop()
    
    while not operators_stack.is_empty():
        postfixed_expression.push(operators_stack.pop())

    print(postfixed_expression)

    prefixed_expression = ArrayStack()
    while not postfixed_expression.is_empty():
        prefixed_expression.push(postfixed_expression.pop())

    number_stack = ArrayStack()
    while not prefixed_expression.is_empty():
        token = prefixed_expression.pop()
        
        if token in operators:
            operand1 = number_stack.pop()
            operand2 = number_stack.pop()
            match token:
                case "+":
                    number_stack.push(operand1 + operand2)
                case "-":
                    number_stack.push(operand2 - operand1)
                case "*":
                    number_stack.push(operand1 * operand2)
                case "/":
                    number_stack.push(operand2 / operand1)
        else:
            number_stack.push(token)
    
    return number_stack.pop()

            
print(calcultate([ "5", "*", "2", "+", "3"]))
```

**Exercício 11:**<br>
Crie um sistema usando uma piha e uma fila para testar se uma determinada string é um palíndromo (ou seja, se os caracteres são lidos da mesma forma, tanto para a frente quanto para a trás)

```python
def is_palindrome(word: str):
    
    midpoint = len(word)//2

    first_half = ArrayQueue(word[:midpoint])
    second_half = ArrayStack(word[midpoint:])

    for i in range(0, midpoint):
        if first_half.dequeue() != second_half.pop():
            return False
        
    return True

if is_palindrome("natan"):
    print("A palavra é um palindromo.")
else:
    print("A palavra NÃO É um palindromo.")
```

**Exercício 12:**<br>
Implemente as classes LinkedStack, LinkedQueue, CircularQueue e LinkedDeque

```python
class LinkedStack:
    class _Node:
        __slots__ = '_element', '_next'
        def __init__(self, element, next_node):
            self._element = element
            self._next = next_node
        
    def __init__(self):
        self._head = None
        self._size = 0
    
    def __len__(self):
        return self._size
    
    def is_empty(self):
        return self._size == 0
    
    def push(self, e):
        self._head = self._Node(e, self._head)
        self._size += 1

    def top(self):
        if self.is_empty():
            raise IndexError('Peek from an empty stack.')
        return self._head._element
    
    def pop(self):
        if self.is_empty():
            raise IndexError('Peek from an empty stack.')
        
        answer = self._head._element

        self._head = self._head._next
        self._size -= 1
        return answer
    
    def __str__(self):
        elements = []
        current_node = self._head
        while current_node:
            elements.append(current_node._element)
            current_node = current_node._next
        
        return str(elements)
    


class LinkedQueue:
    class _Node:
        __slots__ = '_element', '_next'
        def __init__(self, element, next_node):
            self._element = element
            self._next = next_node

    def __init__(self):
        self._head = None
        self._tail = None
        self._size = 0

    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0

    def first(self):
        if self.is_empty():
            raise IndexError('Queue is empty')
        return self._head._element

    def enqueue(self, e):
        new_node = self._Node(e, None)
        if self.is_empty():
            self._head = new_node
        else:
            self._tail._next = new_node
        self._tail = new_node
        self._size += 1

    def dequeue(self):
        if self.is_empty():
            raise IndexError('Queue is empty')
        
        answer = self._head._element
        self._head = self._head._next
        self._size -= 1
        
        if self.is_empty():
            self._tail = None
            
        return answer

    def __str__(self):
        elements = []
        curr = self._head
        while curr:
            elements.append(curr._element)
            curr = curr._next
        return f"Queue: {elements}"
    

class CircularQueue:
    class _Node:
        __slots__ = '_element', '_next'
        def __init__(self, element, next_node):
            self._element = element
            self._next = next_node

    def __init__(self):
        self._tail = None
        self._size = 0

    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0

    def first(self):
        if self.is_empty():
            raise IndexError('Queue is empty')
        head = self._tail._next
        return head._element

    def enqueue(self, e):
        new_node = self._Node(e, None)
        if self.is_empty():
            new_node._next = new_node
            self._tail = new_node
        else:
            new_node._next = self._tail._next 
            self._tail._next = new_node       
            self._tail = new_node             
        self._size += 1

    def dequeue(self):
        if self.is_empty():
            raise IndexError('Queue is empty')
        
        old_head = self._tail._next
        if self._size == 1:
            self._tail = None
        else:
            self._tail._next = old_head._next
            
        self._size -= 1
        return old_head._element

    def rotate(self):
        if self._size > 0:
            self._tail = self._tail._next

    def __str__(self):
        if self.is_empty():
            return "CircularQueue: []"
        elements = []
        head = self._tail._next
        curr = head
        while True:
            elements.append(curr._element)
            curr = curr._next
            if curr is head:
                break
        return f"CircularQueue: {elements}"
```

**Exercicio 13:**<br>
Forneca um algoritmo para encontrar o penultimo no em uma lista encadeada simples na qual o ultimo no e indicado por uma proxima referencia de None


```python
class LinkedList:
    class _Node:
        __slots__ = '_element', '_next'
        def __init__(self, element, next_node):
            self._element = element
            self._next = next_node

    def __init__(self):
        self._head = None
        self._size = 0
    
    def __len__(self):
        return self._size
    
    def is_empty(self):
        return self._size == 0
    
    def append(self, e):
        new_node = self._Node(e, None)
        
        if self.is_empty():
            self._head = new_node
            self._size += 1
        else:
            current_node = self._head
            while not current_node._next == None:
                current_node = current_node._next
            
            current_node._next = new_node
            self._size += 1

    def remove(self, target):
        if self.is_empty():
            raise IndexError('Remove from an empty list.')
        
        current_node = self._head
        if current_node._element == target:
            self._head = current_node._next
            return
        
        while not current_node._next == None:
            if current_node._next._element == target:
                current_node._next = current_node._next._next
                return
            current_node = current_node._next
        
        if current_node._element == target:
            current_node._next = current_node._next._next
            return

    
    def __str__(self):
        elements = []
        current_node = self._head
        while current_node:
            elements.append(current_node._element)
            current_node = current_node._next
        
        return str(elements)
    

def find_pendultimate(linked_list: LinkedList):
    current_node = linked_list._head
    while not current_node._next._next == None:
        
        current_node = current_node._next
    return current_node

linked_list = LinkedList()
linked_list.append(1)
linked_list.append(2)
linked_list.append(3)

# Expected output: 2
print(find_pendultimate(linked_list)._element)
```

**Exercicio 14:**<br>
Descreva um bom algoritmo para concatenar duas listas encadeadas simples L e M, dadas apenas referencias ao primeiro no de cada lista, em uma unica lista L que contem todos <br>
os nos de L seguidos por todos os nos de M

```python
def concatenate_ll(L: LinkedList, M: LinkedList):
    current_node = L._head
    while not current_node._next == None:
        current_node = current_node._next
    current_node._next = M._head


L = LinkedList()
M = LinkedList()

L.append(1)
L.append(2)
L.append(3)

M.append(4)
M.append(5)
M.append(6)

concatenate_ll(L, M)

# Expected output: [1, 2, 3, 4, 5, 6]
print(L)
```

```python
def count_ll_elements(node: LinkedList._Node):
    current_node = node
    if current_node._next != None:
        return 1 + count_ll_elements(current_node._next)
    else:
        return 1


linked_list_2 = LinkedList()
linked_list_2.append(1)
linked_list_2.append(2)
linked_list_2.append(3)

# Expected output: 3
print(count_ll_elements(linked_list_2._head))
```
